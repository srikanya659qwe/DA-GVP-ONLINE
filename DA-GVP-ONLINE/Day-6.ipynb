#### pandas DataFrame:
    - pd.DataFrame([e1,e2,e3...en])

```python
li1=[12,43,45,67,78,9]
df=pd.DataFrame(li1)
df
```

```python
li2=[[12,23,34],[34,45,56],[56,67,78]]
df1=pd.DataFrame(li2)
df1
```

```python
df1.index=['i','j','k']
df1
```

```python
df1.columns=['x','y','z']
df1
```
