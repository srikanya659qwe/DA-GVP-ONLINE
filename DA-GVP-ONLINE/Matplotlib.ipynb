#### lineplot/linear plot

```python
print(dir(plt))
```

```python
x_axis=[67,89,90,12,34]
y_axis=[90,65,43,32,21]
plt.plot(x_axis,y_axis,label="Icon")
plt.title('Linear Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc="upper right")
plt.show()
```

```python
import numpy as np
```

```python
a=np.arange(1,6)
plt.plot(a,a+0.5,color='red')
plt.plot(a,a+1.5,color='green',linestyle=':')
plt.plot(a,a+2.0,color='blue',linestyle='-.')
plt.plot(a,a+2.5,color='yellow',linestyle='--')
plt.plot(a,a+3.0,color='orange',linestyle='-')
```

```python
y=np.linspace(1,20,10)
y
```

```python
x=np.linspace(1,30,50)
plt.plot(x,np.sin(x),color='r',marker='o',markersize=8,markerfacecolor='y') #(o,H,<,>,^)
plt.plot(x,np.cos(x),color='g',marker='H',markersize=8,markerfacecolor='b')
```

#### barplot
    - plt.bar()

```python
z=np.arange(5)
z
```

```python
plt.bar(z,[90,82,55,67,77],width=0.33,color='r')
plt.bar(z+0.33,[23,90,92,44,30],width=0.33,color='g')
```

```python
year=['2020 dec','2021 dec','2022 dec','2023 dec']
gold_price=[40000,50000,60000,70000]
plt.bar(year,gold_price,width=0.4,color=['r','m','pink','green'])
```

### scatter plot():
    - it is a similar to lineplot
    - plt.scatter()

```python
plt.scatter([1,3],[5,6])
```

```python
x1=np.random.randint(100,1000,300)
y1=np.random.randint(500,1500,300)
plt.scatter(x1,y1)
```

```python
x2=np.linspace(0,10,30)
x2
```

```python
y2=np.sin(x2)
plt.scatter(x2,y2,marker='^',color='r')
```

#### boxplot():
        - plt.boxplot()

```python
num=np.random.randint(0,10000,1000)
num
```

```python
plt.boxplot(num)
```

#### piechart
        plt.pie()
        

```python
games=['cricket','chess','ludo','koko']
score=[70,54,45,60]
plt.pie(score,labels=games,labeldistance=1.5,shadow=True,startangle=90)
```

#### histogram():

```python
z1=np.arange(1,35)
plt.hist(z1)
```
